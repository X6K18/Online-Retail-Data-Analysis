# **DATA CLEANING**

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_excel('data/raw/Online Retail.xlsx')
df.head(10)
```

```python
df.info()
```

## **1. Removing Duplicates**

```python
duplicate_values = df.duplicated().sum()
print(f"Number of duplicate values: {duplicate_values}")
```

```python
df = df.drop_duplicates()
duplicate_values_after = df.duplicated().sum()
```

```python
print(f"Number of duplicate values after dropping duplicates: {duplicate_values_after}")
```

```python
df.info()
```

## **2. Handing missing value**

```python
# Check for missing values 
missing_values = df.isnull().sum()
print(f"Missing values in each column:\n{missing_values}")
```

```python
df['CustomerID'] = df['CustomerID'].fillna("Unknown Customer")
```

```python
df['CustomerID'].value_counts()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Description'] = df['Description'].fillna("No Description")
df['Description'].value_counts()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.to_csv('data/processed/online_retail_cleaned_filled.csv', index=False)
```
